```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as bs
import seaborn as sns
import requests 
import re
```

```python
page = requests.get("https://www.dell.com/en-in/search/laptop")
page
```

```python
soup = bs(page.text)
```

```python
Model = []
for i in soup.find_all("h3",class_="ps-title"):
    a = i.text
    Model.append(a)
```

```python
len(Model)
```

```python
Ram = []
for i in soup.find_all("div",class_="short-specs ps-dds-font-icon dds_memory"):
    a = i.text
    b = re.findall("(\d+ GB),",a)
    Ram.append(b[0])
```

```python
len(Ram)
```

```python
Rom = []
for i in soup.find_all("div",class_="short-specs ps-dds-font-icon dds_hard-drive"):
    a = i.text
    b = re.findall("(\d+ GB),",a)
    c = re.findall("(\d+ TB),",a)
    if b:
        Rom.append(b)
    elif c:
        Rom.append(c)
```

```python
len(Rom)
```

```python
RomType = []
for i in soup.find_all("div",class_="short-specs ps-dds-font-icon dds_hard-drive"):
    a = i.text
    b = a.split(",")
    RomType.append(b[-1])
```

```python
len(RomType)
```

```python
Processor = []
for i in soup.find_all("div",class_="short-specs ps-dds-font-icon dds_processor"):
    a = i.text
    Processor.append(a)
```

```python
len(Processor)
```

```python
Price = []
for i in soup.find_all("div",class_="ps-dell-price ps-simplified"):
    a = i.text
    Price.append(a)
```

```python
text = "https://www.dell.com/en-in/search/laptop?p="
```

```python
Model = []
Ram = []
Rom = []
Processor = []
Price = []
RomType = []
for i in range(1,5):
    url = text+str(i)+"&t=Product"
    print(url)
    page = requests.get(url)
    soup = bs(page.text)
    for i in soup.find_all("h3",class_="ps-title"):
        a = i.text
        Model.append(a)
    print(len(Model))
    for i in soup.find_all("div",class_="short-specs ps-dds-font-icon dds_memory"):
        a = i.text
        b = re.findall("(\d+ GB),",a)
        if b:
            Ram.append(b[0])
        else:
            Ram.append(np.NAN)
    print(len(Ram))
    for i in soup.find_all("div",class_="short-specs ps-dds-font-icon dds_hard-drive"):
        a = i.text
        b = re.findall("(\d+ GB),",a)
        c = re.findall("(\d+ TB),",a)
        if b:
            Rom.append(b[0])
        elif c:
            Rom.append(c[0])
    print(len(Rom))
    for i in soup.find_all("div",class_="short-specs ps-dds-font-icon dds_processor"):
        a = i.text
        Processor.append(a)
    print(len(Processor))
    for i in soup.find_all("div",class_="ps-dell-price ps-simplified"):
        a = i.text
        Price.append(a)
    print(len(Price))
    for i in soup.find_all("div",class_="short-specs ps-dds-font-icon dds_hard-drive"):
        a = i.text
        b = a.split(",")
        RomType.append(b[-1])
    print(len(RomType))
```
